# Analysis of Life Expectancy and GDP data of six countries
***

This Data Analysis will be mainly focusing on these questions:
- Has life expectancy increased over time in the six nations?
- Has GDP increased over time in the six nations?
- Is there a correlation between GDP and life expectancy of a country?
- What is the average life expectancy in these nations?
- What is the distribution of that life expectancy?

#### First of all, we need to import the tools needed and load the data set.
***

Tools needed:
- pandas
- numpy
- matplotlib
- seaborn

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

df = pd.read_csv('all_data.csv')
print(df)
```

### Now we can start to answer our questions.
***

#### Has life expectancy increased over time in the six nations?
***
The approach to this questions will be:
1. Get a list of the six countries using `np.unique()`.
2. For every country, I will plot a line graph of life expectancy against the years.

```python
countries = np.unique(df['Country'])
years = np.unique(df['Year'])

plt.figure(figsize=(10, 10))
for i in range(len(countries)):
    ax = plt.subplot(2, 3, i+1)
    x = years
    y = df[df['Country'] == countries[i]]['Life expectancy at birth (years)']
    ax.plot(x, y)
    plt.xlabel('Years')
    plt.ylabel('Life Expectancy')
    plt.title(countries[i])
plt.show()
```

Thus, the above graph shows that life expectancy increased over time in the six nations.
***

#### Has GDP increased over time in the six nations?
***
For this question, the approach is similar to the question above.

```python
plt.figure(figsize=(10, 10))
for i in range(len(countries)):
    ax = plt.subplot(2, 3, i+1)
    x = years
    y = df[df['Country'] == countries[i]]['GDP']
    ax.plot(x, y)
    plt.xlabel('Years')
    plt.ylabel('GDP')
    plt.title(countries[i])
plt.subplots_adjust(wspace=0.8, hspace=0.25)
plt.show()
```

Thus, the above graph shows that there is GDP growth over time in the six nations.
***

#### Is there a correlation between GDP and life expectancy of a country?
***
We will be using a scatter plot to show whether there is a correlation between GDP and life expectancy for six countries.

```python
plt.figure(figsize=(14, 10))
for i in range(len(countries)):
    ax = plt.subplot(2, 3, i+1)
    x = df[df['Country'] == countries[i]]['Life expectancy at birth (years)']
    y = df[df['Country'] == countries[i]]['GDP']
    ax.scatter(x, y)
    z = np.polyfit(x, y, 1)
    p = np.poly1d(z)
    ax.plot(x, p(x))
    plt.xlabel('Life expectancy')
    plt.ylabel('GDP')
    plt.title(countries[i])
plt.subplots_adjust(wspace=0.8, hspace=0.25)
plt.show()
```

This thus shows that there is a positive correlation between GDP and life expectancy for the six countries.
***

#### What is the average life expectancy in these nations?
My approach to this question will be to contruct a bar graph, which x-axis will be the years and y axis will be the average life expectancy. This bar graph will be able to visualize the average life expectancy of these countries throughout the years.

```python
plt.figure(figsize=(20, 30))
sns.barplot(data = df, x='Year', y='Life expectancy at birth (years)', hue='Country')
plt.show()
```

From this graph, we can see that the top 3 countries with the most average life expectancy are Germany, Chile and the US, while the bottom 3 are Mexico, China and Zimbabwee. Zimbabwee notably has a significant lower life expectancy than the other countries.
***

#### What is the distribution of that life expectancy?
My approach to this question is to contruct a violin plot to show the distribution of life expectancy among all countries.

```python
plt.figure(figsize=(10, 10))
sns.violinplot(x=df['Life expectancy at birth (years)'])
plt.show()
```

We can see that the distribution is left skewed, which shows that there is a significant dispairity between the high life expectancy countries(like US and Germany) and the low life expectancy countries(Zimbabwee).
